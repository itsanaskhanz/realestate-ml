```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../data/interim/data_cleaned.csv", parse_dates=["date"])
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
```

```python
df["age_at_sale"] = df["year"] - df["yr_built"]
df["was_renovated"] = df["yr_renovated"].apply(lambda x: 1 if x > 0 else 0)
df["years_since_renovation"] = df.apply(lambda x: x["year"] - x["yr_renovated"] if x["yr_renovated"] > 0 else 0 , axis=1)
df["total_sqft"] = df["sqft_living"] + df["sqft_basement"]
```

```python
cols_to_drop = [
    'date',          
    'year',          
    'street',        
    'statezip',      
    'country',       
    'yr_built',     
    'yr_renovated',   
]
df.drop(columns=cols_to_drop, inplace=True)
```
