```python
import pandas as pd
import numpy as np
import joblib
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
X_test = pd.read_csv("../../data/processed/X_test.csv")
y_test = pd.read_csv("../../data/processed/y_test.csv", header=None).squeeze()
```

```python
best_model = joblib.load('../../models/best_model.pkl')
y_pred = best_model.predict(X_test)
```

```python
if hasattr(best_model, 'best_params_'):
    print(best_model.best_params_)

elif 'ridge' in str(best_model):
    print(best_model.named_steps['ridge'].alpha_)

elif 'lasso' in str(best_model):
    print(best_model.named_steps['lasso'].alpha_)
```

```python
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

final_results = pd.DataFrame({
    "MAE": mae,
    "R2": r2
}, index=[0])
final_results
```

```python
import os
os.makedirs("../../reports", exist_ok=True)

final_results.to_csv('../../reports/final_metrics.csv', index=False)
print("Results saved to reports/final_evaluation.csv")
```
